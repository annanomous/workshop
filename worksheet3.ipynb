3.2 Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of
the passenger(i.e., Fare/Age).

```python
df['Fare_per_Year'] = df.apply(
    lambda row: row['Fare'] / row['Age'] if row['Age'] > 0 else None, axis=1)

print(df[['Fare', 'Age' , 'Fare_per_Year']])
```
